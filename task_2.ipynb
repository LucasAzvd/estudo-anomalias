# Task 2 - Convert JSON Data

```python
df = pd.read_json("data/data.json")
```

```python
df_items = pd.DataFrame()
for i in range(len(df)):
    df_aux_items = pd.DataFrame(df['ItemList'].iloc[i])
    df_aux_items['NFeID'] = df['NFeID'].iloc[i]
    df_items = pd.concat([df_items, df_aux_items])

df_items.head(2)
```

```python
del df['ItemList']
```

Uma alternativa para um problema mais complexo é a utilização de `lambda` functions para extração da informação desejada. Caso esse dado esteja em um banco de dados como Postgres por exemplo, essa extração pode ser feita diretamente no banco de dados, sem a necessidade de trazer todos os dados para a aplicação e processá-los em memória.
